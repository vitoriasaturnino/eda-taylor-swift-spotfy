### 3. Visualização de dados

#### Heatmap

Este mapa de calor funciona por correlação. Isso mostra quais variáveis ​​​​estão correlacionadas entre si em uma escala em que 1 é a mais correlacionada e -1 não está correlacionada de forma alguma.

```python
# Filtra apenas as colunas numéricas
df_numeric = df.select_dtypes(include=['number'])

# Calcula a matriz de correlação entre as colunas numéricas
correlation_matrix = df_numeric.corr()

# Defina uma paleta de cores personalizada
custom_palette = sns.diverging_palette(350, 300, as_cmap=True)

# Formatação do heatmap
plt.figure(figsize=(8, 6))
sns.set(font_scale=1)
sns.heatmap(correlation_matrix, cmap=custom_palette, linewidths=0.5, fmt=".2f", cbar=True)

# Título do heatmap
plt.title('Correlação entre Características Numéricas')

# Exibe o heatmap
plt.show()
```

#### Histplot

Um histograma é uma ferramenta de visualização clássica que representa a distribuição de uma ou mais variáveis ​​contando o número de observações que se enquadram em compartimentos discretos.

```python
# Escolha a coluna que você deseja visualizar (por exemplo, 'popularity')
data_column = df['energy']

# Formatação do histplot
plt.figure(figsize=(8, 4))
sns.set(font_scale=1)

# Cria o histplot
sns.histplot(data_column, bins=20, kde=True, color='purple', edgecolor='black')

# Título do gráfico
plt.title('Distribuição da Energia das Faixas Musicais')

# Rótulos dos eixos
plt.xlabel('energy')
plt.ylabel('loudness')

# Exibe o histplot
plt.show()
```

#### Bar Chart

Representa a média de popularidade de cada álbum do conjunto de dados.

```python
# Agrupe o DataFrame por álbum e calcule a média de popularidade para cada álbum
album_popularity = df.groupby('album')['popularity'].mean().reset_index()

# Ordene os dados por popularidade média
album_popularity = album_popularity.sort_values(by='popularity', ascending=False)

# Formatação do gráfico de barras empilhadas
plt.figure(figsize=(8, 4))
sns.set(font_scale=1)
plt.bar(album_popularity['album'], album_popularity['popularity'], color='pink')

# Título e rótulos dos eixos
plt.title('Popularidade Média por Álbum')
plt.xlabel('Álbum')
plt.ylabel('Popularidade Média')

# Rotação dos rótulos do eixo x para melhor visualização
plt.xticks(rotation=90, fontsize=10)

# Exibe o gráfico
plt.tight_layout()
plt.show()
```

#### Plot

O gráfico de linha mostra a evolução da popularidade média das músicas de Taylor Swift ao longo dos anos, com o ano no eixo x e a popularidade média no eixo y

```python
# Converte a coluna 'release_date' para o tipo de dado datetime
df['release_date'] = pd.to_datetime(df['release_date'])

# Extrai o ano da coluna 'release_date' e cria uma nova coluna 'year'
df['year'] = df['release_date'].dt.year

# Agrupa as músicas por ano e calcula a popularidade média para cada ano
yearly_popularity = df.groupby('year')['popularity'].mean().reset_index()

# Formatação do gráfico
plt.figure(figsize=(8, 4))
sns.set(font_scale=1)
plt.plot(yearly_popularity['year'], yearly_popularity['popularity'], marker='o', linestyle='-',
         color='magenta', markersize=8, label='Popularidade Média')

# Título e rótulos dos eixos
plt.title('Evolução da Popularidade das Músicas de Taylor Swift ao Longo do Tempo')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Popularidade Média')
plt.xticks(rotation=45)

# Legenda
plt.legend()

# Exibe o gráfico
plt.tight_layout()
plt.grid(True)
plt.show()
```
